Lets learn how to be lazy and write less code with loops. 

A loop is a line of code that will run a block of code multiple times. 

‚≠ê The fundamental rule of programming, **"Don't Repeat Yourself"**. This is
known as the **DRY principle**. This means you should not repeat the same blocks
of code in your application. We will use loop help us not to repeat our code.

There are two types of loops. A for-loop and a while-loop. We will learn about both.

# For loop

Let's look at the code from the *02_operations* notebook. In the code below we are make
squirtle draw a 4 sided shape. Notice how we repeat the same 3 lines of code. This does not follow the **DRY principle**

```python
squirtle = Turtle()

color = 'red'
degrees = round(360/4)
squirtle.pencolor(color)

color = 'yellow'
```

Below is the repeated 3-line block of code. We repeat the following block of
code 4 times:

```python
degrees = round(360/4)
squirtle.pencolor(color)
squirtle.circle(80, degrees)
```

Let's clean up our code with a for-loop.

A for-loop runs a block of code over and over again with a different value. Each
time the loop runs, a new value is given to it by a collection of values.

Let's first create a collection known as a `list`:

```python
colors = ["red", "green", "blue", "yellow"]
```

In the code above, we've create a `list` of `string`. Each string is a color. 

Now that we have a collection let's create a for-loop:

```python
colors = ["red", "green", "blue", "yellow"]

for color in colors:
    ...
```
We create the for-loop by starting with the `for` keyword. Then creating a
variable. The variable is named "color". The variable is where the value from
the colors `list` will be put each time we run the loop. Finally, we need to end
the loop with the `in` keyword and the name of a collection. 

Now we need to add the repeated block of code to the loop.

```python
colors = ["red", "green", "blue", "yellow"]

for color in colors:
    degrees = round(360/4)
    squirtle.pencolor(color)
    squirtle.circle(80, degrees)
```

It is very important that the code block has 4 spaces in front of the lines
under the for-loop. The spaces tell python that the lines of code belong to the
for-loop.

The for loop will run the same number of times there are colors in the colors
list. Each time through the loop the color variable will be set to be equals to
one of the colors in the colors list.

Run the code below to try out the loop. Try changing the colors in the list. Try
adding a color to the colors list. What happen?



````python
from mobilechelonian import Turtle

squirtle = Turtle()

colors = ["red", "green", "blue", "yellow"]

for color in colors:
    degrees = round(360/4)
    squirtle.pencolor(color)
    squirtle.circle(80, degrees)
````
